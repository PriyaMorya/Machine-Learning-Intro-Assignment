Question 1:Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS)
- 1. **Artificial Intelligence (AI)**

 - Definition: The broadest field. AI is about creating systems that can perform tasks that normally require human intelligence.

 - Scope: Includes reasoning, planning, problem-solving, language understanding, perception, and decision-making.

 - Examples:

  Rule-based expert systems (before ML became dominant)

   Chatbots

 Self-driving cars
 - Key Point: AI is the umbrella term — ML, DL, and parts of DS fit inside it

2. **Machine Learning (ML)**

 - Definition: A subset of AI focused on algorithms that learn from data instead of being explicitly programmed.

 - How it works: The system finds patterns in historical data and uses them to make predictions or decisions.

 - Examples:

  Spam email filtering

  Predicting house prices

 Recommendation engines (Netflix, Amazon)

 - Key Point: ML is the current dominant approach to building AI systems.

3. **Deep Learning (DL)**

 - Definition: A specialized subset of ML that uses artificial neural networks with many layers ("deep" networks).

 - Why special?: DL can automatically learn complex features from raw data (e.g., images, audio, text) without manual feature engineering.

 - Examples:

   Image recognition (e.g., detecting cats in photos)

   Speech-to-text (Siri, Alexa)

- Key Point: DL is a technique within ML that powers today’s most advanced AI systems.

4. **Data Science (DS)**

- Definition: An interdisciplinary field about extracting insights and knowledge from data — combining statistics, programming, ML, and domain expertise.

 - Goal: To understand, visualize, and solve business or scientific problems using data.

 - Examples:

   Analyzing customer behavior to improve sales

   Finding trends in healthcare data

   Running A/B tests to guide product decisions

 - Key Point: DS is broader than AI/ML — it includes data cleaning, exploration, visualization, reporting, and sometimes building ML models.



Question 2: What are the types of machine learning? Describe each with one
real-world example.
- The three main types of machine learning are :
1. **Supervised Learning**
 - What it is:
The algorithm is trained on a labeled dataset — meaning each input has a corresponding correct output.

 - Goal: Learn the mapping between inputs (features) and outputs (labels), then make predictions on new, unseen data.

 - Tasks:

  Classification (predict categories, e.g., spam/not spam)

  Regression (predict continuous values, e.g., house prices)

 - Example:
Email Spam Detection: The model is trained on past emails labeled as spam or not spam. It then classifies new emails accordingly.

2. **Unsupervised Learning**

- What it is:
The algorithm is trained on an unlabeled dataset — only inputs are given, no explicit outputs.

- Goal: Discover hidden patterns, structures, or relationships in the data.

- Tasks:

  Clustering (group similar data points)

  Dimensionality Reduction (reduce complexity, e.g., PCA)

- Example:

  Customer Segmentation in Marketing: A retailer groups customers into clusters (e.g., budget shoppers, premium buyers) based on purchase history, without predefined labels.

3. **Semi-Supervised Learning**

 - What it is:
A mix of supervised and unsupervised learning — the algorithm is trained on a small amount of labeled data plus a large amount of unlabeled data.

 - Goal: Improve learning accuracy when labeling data is expensive or time-consuming.

 - Example:

  Medical Diagnosis with Few Labels: Doctors label a small set of medical images (e.g., cancer vs. non-cancer), and the model uses these plus many unlabeled images to learn better.

4. **Reinforcement Learning (RL)**

 - What it is:
The algorithm learns by interacting with an environment. It takes actions, receives rewards/penalties, and improves strategy over time.

 - Goal: Maximize cumulative reward by learning an optimal policy (what action to take in each situation).

 - Example:

  Self-Driving Cars: The car’s AI learns to drive by trial and error — actions (steering, braking, accelerating) are rewarded if they lead to safe and efficient driving, penalized if they cause accidents or inefficiency.




Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.
- 1. **Overfitting**

    - Definition: When a model learns the training data too well, including its noise and random fluctuations, instead of the true underlying patterns.

   - Result: The model performs very well on training data but poorly on unseen (test) data because it fails to generalize.

   - Analogy: Like a student memorizing past exam questions without understanding the concepts — they do well on practice tests but fail on new ones.

  - Example: A decision tree that grows too deep and perfectly classifies the training data but makes wrong predictions on new inputs.

2. **Underfitting**

   - Definition: When a model is too simple to capture the underlying structure of the data.

   - Result: It performs poorly on both training and test data.

   - Analogy: Like a student who skimmed the textbook but didn’t study enough — they can’t even do well on practice problems.

   - Example: Using a straight line (linear regression) to model a highly nonlinear relationship.

3. **Bias-Variance Tradeoff**

   - Bias: Error from overly simplistic assumptions in the model (leads to underfitting).

   - Variance: Error from too much sensitivity to training data (leads to overfitting).

   - Tradeoff:

   A model with high bias, low variance → underfits.

   A model with low bias, high variance → overfits.

   - The goal is to find the sweet spot: balanced bias and variance for the best generalization.


   - Example:

   Linear regression on complex data → high bias, underfit.

   A very deep neural network on small data → high variance, overfit.

   A moderately regularized model → balanced, generalizes well.


Question 4: What are outliers in a dataset, and list three common techniques for
handling them.
- **Definition:** Outliers are data points that are significantly different from the majority of the dataset(extreme points).
Causes:

Measurement or data entry errors (e.g., typing “1000” instead of “100”).

Natural variation (e.g., extremely tall or short individuals in height data).

Rare but valid events (e.g., a sudden stock market crash).

Impact:

Can bias averages (mean, standard deviation).

May lead to incorrect model training, especially in algorithms sensitive to distance (like k-means, regression).

**Three Common Techniques for Handling Outliers**

1. *Removal (Filtering)*

 - What: Delete outliers if they are errors or irrelevant to the problem.

 - When to use: Only when you’re confident they are mistakes, not meaningful data.

 - Example: Removing negative ages from a dataset of patients.


2. *Transformation / Capping*

 - What: Apply transformations to reduce the impact of outliers.

 - Log or square root transformation: reduces skewness.

 - Winsorization (capping): Replace extreme values with a certain percentile (e.g., values above the 95th percentile are set to the 95th percentile).

 - Example: Applying a log transform on highly skewed income data to reduce the effect of billionaires.

3. *Imputation / Replacement*

 - What: Replace outliers with more reasonable values such as:

 - Mean/median of the data.

 - Nearest neighbor values.

 - When to use: When outliers are likely due to noise or errors but keeping the sample size is important.

 - Example: Replacing a mistyped "999" temperature reading with the average of nearby valid readings.


Question 5: Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.
- The Process of Handling Missing Values

1. **Identify Missing Values**

   - Use tools like .isnull() in pandas or summary statistics to check missingness.

   - Visualize patterns (e.g., heatmaps, missing value plots).

2. **Understand the Cause**

   - MCAR (Missing Completely at Random) → No pattern, safe to handle with imputation or removal.

   - MAR (Missing at Random) → Missingness depends on other variables.

   - MNAR (Missing Not at Random) → Missingness depends on the value itself (trickiest).

3. **Decide on Strategy**

    - Remove: Drop rows/columns if missingness is small and random.

    - Impute: Replace missing values with estimated values.

    - Flag: Add a binary feature indicating missingness (useful in ML).

# **Imputation Techniques**

**For Numerical Data**

 *Mean/Median Imputation*

   - Replace missing values with the column’s mean or median.

   - Best when: Data is not highly skewed (mean) or is skewed (median).

  - Example: If some salaries are missing, replace them with the median salary.


**For Categorical Data**

 *Mode Imputation*

  - Replace missing values with the most frequent category (mode).

  - Best when: A few categories dominate the distribution.

  - Example: If “gender” has missing entries, replace with the most common value.

Question 8: Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.randn(1000)

# Step 2: Introduce 50 random missing values
data_missing = data.copy()
missing_indices = np.random.choice(1000, 50, replace=False)
data_missing[missing_indices] = np.nan

# Step 3: Fill missing values with the column mean
data_filled = pd.Series(data_missing).fillna(np.nanmean(data_missing))

# Step 4: Plot histograms before and after imputation
plt.figure(figsize=(12, 5))

# Before imputation
plt.subplot(1, 2, 1)
plt.hist(data_missing[~np.isnan(data_missing)], bins=30, edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1, 2, 2)
plt.hist(data_filled, bins=30, edgecolor='black', color='orange')
plt.title("After Imputation (Mean)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

2) Runnable Python script (demonstration)

The script below creates a synthetic retail transactions dataset, demonstrates each preparation step (imputation, outlier handling, oversampling, encoding, scaling) and prints intermediate results.

# (Code executed below in this notebook; shown here again for clarity)

```python
# (Code executed below in this notebook; shown here again for clarity)

import numpy as np
import pandas as pd
from sklearn.utils import resample
from sklearn.preprocessing import StandardScaler
from collections import Counter

np.random.seed(42)
n = 1000

# --- Synthetic dataset ---
ages = np.random.normal(35, 12, size=n)
ages = np.clip(ages, 18, 90)
missing_idx = np.random.choice(n, size=int(0.15 * n), replace=False)
ages[missing_idx] = np.nan

transaction_amount = np.random.exponential(scale=50, size=n) + 5
outlier_idx = np.random.choice(n, size=10, replace=False)
transaction_amount[outlier_idx] = transaction_amount[outlier_idx] * 20

payment_method = np.random.choice(
    ['card', 'cash', 'paypal', 'bank_transfer', 'mobile_pay'],
    size=n,
    p=[0.5, 0.2, 0.15, 0.1, 0.05]
)

num_items = np.random.poisson(2, size=n) + 1
is_fraud = np.random.choice([0, 1], size=n, p=[0.98, 0.02])

df = pd.DataFrame({
    'age': np.round(ages, 1),
    'transaction_amount': np.round(transaction_amount, 2),
    'payment_method': payment_method,
    'num_items': num_items,
    'is_fraud': is_fraud
})

# --- Initial checks ---
print("Initial dataset preview:\n", df.head())
print("\nMissing values per column:\n", df.isnull().sum())
print("\nInitial class distribution:\n", df['is_fraud'].value_counts())

# --- Missing ages: flag + median imputation ---
median_age = df['age'].median()
df['age_missing_flag'] = df['age'].isnull().astype(int)
df['age_imputed'] = df['age'].fillna(median_age)
print("\nMedian age used for imputation:", median_age)
print("Missing after imputation:", df['age_imputed'].isnull().sum())

# --- Outliers: detect with IQR, winsorize, and log transform ---
Q1 = df['transaction_amount'].quantile(0.25)
Q3 = df['transaction_amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_mask = (df['transaction_amount'] < lower_bound) | (df['transaction_amount'] > upper_bound)

lower_pct = df['transaction_amount'].quantile(0.01)
upper_pct = df['transaction_amount'].quantile(0.99)
df['trans_winsor'] = df['transaction_amount'].clip(lower=lower_pct, upper=upper_pct)
df['trans_log1p'] = np.log1p(df['transaction_amount'])

print("\nIQR outlier thresholds: lower={:.2f}, upper={:.2f}".format(lower_bound, upper_bound))
print("Number of IQR-detected outliers:", outliers_mask.sum())
print("\nTransaction amount stats (original / winsorized / log1p):")
print(df[['transaction_amount', 'trans_winsor', 'trans_log1p']].describe().round(2))

# --- Imbalance: simple oversample minority to match majority ---
print("\nClass distribution before balancing:", Counter(df['is_fraud']))
df_majority = df[df['is_fraud'] == 0]
df_minority = df[df['is_fraud'] == 1]
df_minority_upsampled = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42)
df_balanced = pd.concat([df_majority, df_minority_upsampled]).sample(frac=1, random_state=42).reset_index(drop=True)
print("Class distribution after oversampling:", Counter(df_balanced['is_fraud']))

# --- Encode categorical(s) ---
df_encoded = pd.get_dummies(df_balanced, columns=['payment_method'], drop_first=False)

# --- Scale numeric features ---
num_cols = ['age_imputed', 'trans_winsor', 'num_items']
scaler = StandardScaler()
df_encoded[[c + '_scaled' for c in num_cols]] = scaler.fit_transform(df_encoded[num_cols])

print("\nFinal processed dataset preview (first rows):")
print(df_encoded.head().T)
print("\nFinal shape:", df_encoded.shape)
```
